## Part 2: Customer Analysis

#### Q1. What is the Average Age of Customers?

```python
df['age'].mean()
```

### Insight

The average age of customers is approximately 54 years, indicating that the customer base is dominated by middle-aged and senior individuals.

#### Q2. What is the Average Annual Income ?

```python
df['annual_income'].mean()
```

### Insight

The average annual income is approximately 109,743. This suggests that customers belong to a wide range of income categories.

#### Q3. Which Gender Spends More on Average?

```python
df.groupby('gender')['spending_score'].mean().reset_index()
```

### Insight

Male customers have a slightly higher average spending score than female customers. However, the difference is very small, indicating similar spending behavior across genders.

#### Q4. Which Age Group Has the Highest Spending Score?

```python
bins = [0,20,30,40,50,60,100]

labels = ['0-20','21-30','31-40','41-50','51-60','60+']

df['age_group'] = pd.cut(df['age'],bins=bins,labels=labels)

df.groupby('age_group')['spending_score'].mean().reset_index()
```

### Insight

The average spending score is nearly identical across all age groups (50.27–51.40), indicating that age has little influence on customer spending behavior in this dataset. Since no age group shows significantly higher or lower spending, age alone is not a strong factor for customer segmentation or targeted marketing.

#### Q5. Is There Any Relationship Between Annual Income and Spending Score?

```python
df['annual_income'].corr(df['spending_score'])
```

### Insight

The correlation coefficient is very close to zero, indicating almost no relationship between annual income and spending score. Income alone does not explain customer spending behavior.

## Part 3: Customer Segmentation

#### Creating Customer Segments

```python
income_threshold = df['annual_income'].median()

spending_threshold = df['spending_score'].median()

def segment(row):

    if row['annual_income'] >= income_threshold and row['spending_score'] >= spending_threshold:
        return "High Income - High Spending"

    elif row['annual_income'] >= income_threshold and row['spending_score'] < spending_threshold:
        return "High Income - Low Spending"

    elif row['annual_income'] < income_threshold and row['spending_score'] >= spending_threshold:
        return "Low Income - High Spending"

    else:
        return "Low Income - Low Spending"

df['segment'] = df.apply(segment, axis=1)
```

### Insight

The customer base can be divided into four meaningful groups based on income and spending behavior. These segments help businesses design targeted marketing strategies and improve customer engagement.
